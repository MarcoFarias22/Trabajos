```python
# Ejercicio 1: Clase Persona

class Persona:
    def __init__(self, nombre, edad, email=None):
        """Constructor que inicializa los atributos básicos."""
        self.nombre = nombre
        self.edad = edad
        self.email = email

    def saludar(self):
        """Retorna un saludo con el nombre de la persona."""
        return f"Hola, soy {self.nombre}"

    def es_mayor(self):
        """Verifica si la persona tiene 18 años o más."""
        return self.edad >= 18

    def cumplir_anios(self):
        """Incrementa la edad en 1."""
        self.edad += 1

    def __str__(self):
        """Representación legible del objeto en formato string."""
        return f"Persona({self.nombre}, {self.edad})"

if __name__ == "__main__":
    
    usuario = Persona(nombre="Carlos", edad=17)

    print(f"Nombre: {usuario.nombre}")
    print(usuario.saludar())

    print(f"¿Es mayor de edad?: {usuario.es_mayor()}")

    print("Celebrando cumpleaños...")
    usuario.cumplir_anios()

    print(f"Nueva edad: {usuario.edad}")
    print(f"¿Es mayor de edad ahora?: {usuario.es_mayor()}")

    print(usuario)
```

```python
# Ejercicio 2: Clase Producto

class Producto:
    def __init__(self, nombre, precio, stock=0):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock

    def aplicar_descuento(self, porcentaje):
        """Reduce el precio según el porcentaje y lo devuelve."""
        self.precio = self.precio * (1 - porcentaje / 100)
        return self.precio

    def hay_stock(self):
        """Retorna True si hay al menos una unidad disponible."""
        return self.stock > 0

    def vender(self, cantidad=1):
        """
        Reduce el stock si hay unidades suficientes.
        Retorna True si la venta fue exitosa, False de lo contrario.
        """
        if cantidad <= self.stock:
            self.stock -= cantidad
            return True
        else:
            return False

    def __str__(self):
        """Retorna el formato: Nombre $Precio (Unidades u)"""
        return f"{self.nombre} ${self.precio} ({self.stock}u)"


nb = Producto("Notebook", 850000, 15)
print(f"Producto inicial: {nb}")

nuevo_precio = nb.aplicar_descuento(10)
print(f"Precio con descuento (10%): ${nuevo_precio}")

exito_venta = nb.vender(3)
print(f"Venta de 3 unidades exitosa: {exito_venta}")
print(f"Estado tras venta: {nb}")

print("-" * 30)

tk = Producto("Teclado", 45000, 0)
print(f"Producto inicial: {tk}")
intentar_vender = tk.vender(1)
print(f"¿Se pudo vender el teclado?: {intentar_vender}")
```

```python
# Ejercicio 3: Clase Auto

class Auto:
    def __init__(self, marca, modelo, km=0):
        self.marca = marca
        self.modelo = modelo
        self.km = km
        self.encendido = False  

    def encender(self):
        """Cambia el estado del auto a True."""
        self.encendido = True
        print(f"El {self.marca} se ha encendido.")

    def apagar(self):
        """Cambia el estado del auto a False."""
        self.encendido = False
        print(f"El {self.marca} se ha apagado.")

    def conducir(self, distancia):
        """Suma distancia a los km solo si el auto está encendido."""
        if self.encendido:
            self.km += distancia
            print(f"Conduciendo {distancia}km...")
        else:
            print(f"No se puede conducir: El {self.marca} está apagado.")

    def __str__(self):
        """Representación: Marca Modelo (km)"""
        return f"{self.marca} {self.modelo} ({self.km}km)"


mi_auto = Auto("Toyota", "Corolla")
print(f"Estado inicial: {mi_auto}")

mi_auto.conducir(100)
print(f"Kilometraje actual: {mi_auto.km}km") 

mi_auto.encender()
mi_auto.conducir(150)
print(f"Kilometraje tras viaje: {mi_auto.km}km")

mi_auto.apagar()
mi_auto.conducir(50)
print(f"Estado final: {mi_auto}") 
```

```python
# Ejercicio 4: Clase CuentaBancaria

class CuentaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, monto):
        """Suma al saldo solo si el monto es positivo."""
        if monto > 0:
            self.saldo += monto
            print(f"Depósito exitoso: +${monto}")
        else:
            print("Error: El monto a depositar debe ser positivo.")

    def extraer(self, monto):
        """
        Resta del saldo si el monto es positivo y hay fondos suficientes.
        Retorna True si fue exitoso, False de lo contrario.
        """
        if 0 < monto <= self.saldo:
            self.saldo -= monto
            return True
        else:
            return False

    def consultar_saldo(self):
        """Retorna el saldo actual."""
        return self.saldo

    def __str__(self):
        """Representación: Titular: $Saldo"""
        return f"{self.titular}: ${self.saldo}"

cuenta = CuentaBancaria("Ana", 10000)
print(f"Estado inicial -> {cuenta}")


cuenta.depositar(5000)
print(f"Saldo tras depósito: ${cuenta.consultar_saldo()}") 


pudo_extraer = cuenta.extraer(3000)
print(f"¿Extracción de 3000 exitosa?: {pudo_extraer}")
print(f"Saldo actual: ${cuenta.saldo}") 


intento_fallido = cuenta.extraer(50000)
print(f"¿Extracción de 50000 exitosa?: {intento_fallido}")
print(f"Saldo tras intento fallido: ${cuenta.saldo}") 


cuenta.depositar(-100)
print(f"Estado final -> {cuenta}") 
```

```python
# Ejercicio 5: Clase Estudiante

class Estudiante:
    def __init__(self, nombre, legajo, materias=None):
        self.nombre = nombre
        self.legajo = legajo
       
        if materias is None:
            self.materias = []
        else:
            self.materias = materias

    def inscribir(self, materia):
        """Agrega la materia si el estudiante no está ya en ella."""
        if materia not in self.materias:
            self.materias.append(materia)
        else:
            print(f"Aviso: El estudiante ya está inscripto en {materia}.")

    def cantidad_materias(self):
        """Retorna el tamaño de la lista de materias."""
        return len(self.materias)

    def cursa(self, materia):
        """Verifica si la materia existe en la lista."""
        return materia in self.materias

    def __str__(self):
        """Retorna Nombre (Leg: 123) - X materias."""
        n = self.cantidad_materias()
        return f"{self.nombre} (Leg: {self.legajo}) - {n} materias"


est = Estudiante("Juan", 12345)


est.inscribir("Programacion II")
est.inscribir("Matematica")


print(f"¿Cursa Programación II?: {est.cursa('Programacion II')}") # True


est.inscribir("Programacion II")


print(f"Cantidad de materias: {est.cantidad_materias()}") # Debería ser 2


print(est)
```

```python
# Ejercicio 6: Concesionaria de Autos

class Auto:
    def __init__(self, marca, modelo, km=0):
        self.marca = marca
        self.modelo = modelo
        self.km = km
        self.encendido = False

    def encender(self):
        self.encendido = True

    def apagar(self):
        self.encendido = False

    def conducir(self, distancia):
        if self.encendido:
            self.km += distancia
        else:
            print(f"No se puede conducir: El {self.marca} está apagado.")

    def __str__(self):
        return f"{self.marca} {self.modelo} ({self.km}km)"


class Concesionaria:
    def __init__(self, nombre):
        self.nombre = nombre
        self.autos = []

    def agregar_auto(self, auto):
        """Recibe un objeto de la clase Auto y lo agrega a la lista."""
        self.autos.append(auto)

    def buscar_por_marca(self, marca):
        """Retorna una lista con los autos que coincidan con la marca."""
       
        return [a for a in self.autos if a.marca.lower() == marca.lower()]

    def auto_con_mas_km(self):
        """Retorna el objeto Auto que tenga el kilometraje más alto."""
        if not self.autos:
            return None

        mayor_auto = self.autos[0]
        for auto in self.autos:
            if auto.km > mayor_auto.km:
                mayor_auto = auto
        return mayor_auto

    def cantidad_autos(self):
        """Retorna el total de autos en la lista."""
        return len(self.autos)

    def __str__(self):
        """Representación de la concesionaria."""
        return f"Concesionaria {self.nombre}: {self.cantidad_autos()} autos"

mi_agencia = Concesionaria("Ruedas del Sur")


mi_agencia.agregar_auto(Auto("Toyota", "Corolla", 15000))
mi_agencia.agregar_auto(Auto("Ford", "Mustang", 5000))
mi_agencia.agregar_auto(Auto("Toyota", "Etios", 45000))
mi_agencia.agregar_auto(Auto("Fiat", "Cronos", 120000))


print(mi_agencia)
print("-" * 30)


print("Búsqueda de la marca Toyota:")
toyotas = mi_agencia.buscar_por_marca("Toyota")
for t in toyotas:
    print(f" -> {t}")

print("-" * 30)


mas_caminado = mi_agencia.auto_con_mas_km()
print(f"El auto con más kilómetros es: {mas_caminado}")
```
