```python
# Ejercicio 1: Pila con nodos

class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.siguiente = None

# Implementación de la Pila
class Pila:
    def __init__(self):
        self.cima = None

    def es_vacia(self):
        return self.cima is None

    def push(self, dato):
        """Agrega un elemento al tope de la pila."""
        nuevo_nodo = Nodo(dato)
        nuevo_nodo.siguiente = self.cima
        self.cima = nuevo_nodo

    def pop(self):
        """Remueve y retorna el elemento del tope."""
        if self.es_vacia():
            return "La pila está vacía"

        valor = self.cima.dato
        self.cima = self.cima.siguiente
        return valor

    def tope(self):
        """Muestra el elemento superior sin removerlo."""
        if self.es_vacia():
            return "La pila está vacía"
        return self.cima.dato

    def __str__(self):
        """Muestra el contenido de la pila desde el tope a la base."""
        if self.es_vacia():
            return "Pila[]"

        elementos = []
        actual = self.cima
        while actual:
            elementos.append(str(actual.dato))
            actual = actual.siguiente
        return "Pila[" + " -> ".join(elementos) + "]"

mi_pila = Pila()
mi_pila.push(5)
mi_pila.push(3)
mi_pila.push(8)

print(f"Estado inicial: {mi_pila}")
print(f"Pop realizado: {mi_pila.pop()}") # Debería sacar el 8
print(f"Tope actual: {mi_pila.tope()}")   # Debería mostrar el 3
print(f"Estado final: {mi_pila}")
```

```python
# Ejercicio 2: Cola con nodos

class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.siguiente = None

class Cola:
    def __init__(self):
        self.frente = None
        self.final = None

    def es_vacia(self):
        return self.frente is None

    def enqueue(self, dato):
        """Agrega un elemento al final de la cola."""
        nuevo_nodo = Nodo(dato)
        if self.es_vacia():
            self.frente = nuevo_nodo
            self.final = nuevo_nodo
        else:
            self.final.siguiente = nuevo_nodo
            self.final = nuevo_nodo

    def dequeue(self):
        """Remueve y retorna el elemento del frente (FIFO)."""
        if self.es_vacia():
            return "La cola está vacía"

        valor = self.frente.dato
        self.frente = self.frente.siguiente

        if self.frente is None:
            self.final = None

        return valor

    def frente_cola(self):
        """Ver el primer elemento sin removerlo."""
        if self.es_vacia():
            return "La cola está vacía"
        return self.frente.dato

    def __str__(self):
        """Mostrar contenido de frente a cola."""
        if self.es_vacia():
            return "Cola[]"

        elementos = []
        actual = self.frente
        while actual:
            elementos.append(str(actual.dato))
            actual = actual.siguiente
        return "Cola[" + " -> ".join(elementos) + "]"

mi_cola = Cola()
mi_cola.enqueue(1)
mi_cola.enqueue(2)
mi_cola.enqueue(3)

print(f"Estado inicial: {mi_cola}")
print(f"Dequeue realizado: {mi_cola.dequeue()}") # Debería sacar el 1
print(f"Frente actual: {mi_cola.frente_cola()}") # Debería mostrar el 2
print(f"Estado final: {mi_cola}")
```

```python
# Ejercicio 3: Parentesis balanceados

def verificar_parentesis(expresion):
    pila = Pila()
    pares = {')': '(', ']': '[', '}': '{'}

    for caracter in expresion:
        if caracter in '([{':
            pila.push(caracter)

        elif caracter in ')]}':
            if pila.es_vacia():
                return False

            if pila.pop() != pares[caracter]:
                return False

    return pila.es_vacia()

test1 = "(a + b) * [c + {d}]"
test2 = "(a + b] * [c]"
test3 = "((()))"

print(f"Prueba 1 '{test1}': {verificar_parentesis(test1)}") # True
print(f"Prueba 2 '{test2}': {verificar_parentesis(test2)}") # False
print(f"Prueba 3 '{test3}': {verificar_parentesis(test3)}") # True
```

```python
# Ejercicio 4: Simulador de cola

# Usar la clase Cola implementada

def simulador_banco():
    banco = Cola()

    print("--- Llegada de los primeros 5 clientes ---")
    for i in range(1, 6):
        print(f"Cliente {i} entrando a la fila...")
        banco.enqueue(f"Cliente {i}")

    print(f"\nEstado actual de la fila: {banco}\n")

    print("--- Atendiendo a los primeros 3 clientes ---")
    for _ in range(3):
        atendido = banco.dequeue()
        print(f"Atendiendo {atendido}")

    print(f"\nFila después de las primeras atenciones: {banco}\n")

    print("--- Llegan 2 clientes más ---")
    for i in range(6, 8):
        print(f"Cliente {i} entrando a la fila...")
        banco.enqueue(f"Cliente {i}")

    print(f"\nEstado de la fila: {banco}\n")

    print("--- Atendiendo al resto de la fila ---")
    while not banco.es_vacia():
        atendido = banco.dequeue()
        print(f"Atendiendo {atendido}")

    print(f"\nSimulación terminada. {banco}")

simulador_banco()
```

```python
# Ejercicio 5: Decimal a binario

def decimal_a_binario(n):
    if n == 0:
        return "0"

    pila_residuos = Pila()

    while n > 0:
        residuo = n % 2
        pila_residuos.push(residuo)
        n = n // 2

    resultado_binario = ""
    while not pila_residuos.es_vacia():
        resultado_binario += str(pila_residuos.pop())

    return resultado_binario

print(f"Decimal 10 a binario: {decimal_a_binario(10)}")
print(f"Decimal 7 a binario: {decimal_a_binario(7)}")
print(f"Decimal 0 a binario: {decimal_a_binario(0)}")
```

```python
# Ejercicio 6: Cola de prioridad

class NodoPrioridad:
    def __init__(self, dato, prioridad):
        self.dato = dato
        self.prioridad = prioridad
        self.siguiente = None

class ColaPrioridad:
    def __init__(self):
        self.frente = None

    def es_vacia(self):
        return self.frente is None

    def enqueue(self, dato, prioridad):
        """Agrega un elemento en la posición correcta según su prioridad."""
        nuevo_nodo = NodoPrioridad(dato, prioridad)

        if self.es_vacia() or prioridad < self.frente.prioridad:
            nuevo_nodo.siguiente = self.frente
            self.frente = nuevo_nodo
        else:
            actual = self.frente
            while actual.siguiente is not None and actual.siguiente.prioridad <= prioridad:
                actual = actual.siguiente

            nuevo_nodo.siguiente = actual.siguiente
            actual.siguiente = nuevo_nodo

    def dequeue(self):
        """Remueve el elemento de mayor prioridad (el primero de la cola)."""
        if self.es_vacia():
            return "La cola de prioridad está vacía"

        valor = self.frente.dato
        self.frente = self.frente.siguiente
        return valor

    def __str__(self):
        if self.es_vacia():
            return "ColaPrioridad[]"

        elementos = []
        actual = self.frente
        while actual:
            elementos.append(f"({actual.dato}, P:{actual.prioridad})")
            actual = actual.siguiente
        return " -> ".join(elementos)

cola_p = ColaPrioridad()

cola_p.enqueue("tarea1", 3)
cola_p.enqueue("tarea2", 1)
cola_p.enqueue("tarea3", 2)

print(f"Estado de la cola: {cola_p}")

print(f"Atendiendo: {cola_p.dequeue()}")
print(f"Atendiendo: {cola_p.dequeue()}")

print(f"Estado final: {cola_p}")
```
