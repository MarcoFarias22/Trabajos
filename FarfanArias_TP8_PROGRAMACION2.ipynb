```python
# Ejercicio 1: Factorial recursivo
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
# Pruebas
    
print(f"factorial(5) -> {factorial(5)}") # Debería dar 120
print(f"factorial(0) -> {factorial(0)}") # Debería dar 1
print(f"factorial(7) -> {factorial(7)}") # Debería dar 5040
```

```python
# Ejercicio 2: Fibonacci recursivo
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Pruebas

print(f"fibonacci(10) -> {fibonacci(10)}") # Debería dar 55
print(f"fibonacci(6) -> {fibonacci(6)}")   # Debería dar 8
print(f"fibonacci(0) -> {fibonacci(0)}")   # Debería dar 0
```

```python
# Ejercicio 3: Potencia recursiva
def potencia(base, exponente):
    if exponente == 0:
        return 1
    else:
        return base * potencia(base, exponente - 1)

# Pruebas

print(f"potencia(2, 8) -> {potencia(2, 8)}")   # Debería dar 256
print(f"potencia(5, 3) -> {potencia(5, 3)}")   # Debería dar 125
print(f"potencia(10, 0) -> {potencia(10, 0)}") # Debería dar 1
```

```python
# Ejercicio 4: Suma de dígitos
def suma_digitos(n):
    if n < 10:
        return n
    else:
        ultimo_digito = n % 10
        resto_del_numero = n // 10
        return ultimo_digito + suma_digitos(resto_del_numero)

# Pruebas

print(f"suma_digitos(123) -> {suma_digitos(123)}")   # 1+2+3 = 6
print(f"suma_digitos(9999) -> {suma_digitos(9999)}") # 9+9+9+9 = 36
print(f"suma_digitos(0) -> {suma_digitos(0)}")       # 0
```

```python
# Ejercicio 5: Busqueda en lista recursiva
def buscar_en_lista(lista, valor, indice=0):
    if indice == len(lista):
        return -1

    if lista[indice] == valor:
        return indice

    return buscar_en_lista(lista, valor, indice + 1)

# Pruebas

print(f"buscar_en_lista([1, 2, 3, 4, 5], 3) -> {buscar_en_lista([1,2,3,4,5], 3)}") # 2
print(f"buscar_en_lista([1, 2, 3, 4, 5], 10) -> {buscar_en_lista([1,2,3,4,5], 10)}") # -1
print(f"buscar_en_lista([], 5) -> {buscar_en_lista([], 5)}") # -1
```

```python
# Ejercicio 6: Invertir cadena recursiva
def invertir_cadena(s):
    if len(s) <= 1:
        return s
    else:
        ultimo_caracter = s[-1]
        resto = s[:-1]
        return ultimo_caracter + invertir_cadena(resto)

# Pruebas

print(f'invertir_cadena("hola") -> "{invertir_cadena("hola")}"') # "aloh"
print(f'invertir_cadena("a")    -> "{invertir_cadena("a")}"')    # "a"
print(f'invertir_cadena("")     -> "{invertir_cadena("")}"')     # ""
```

```python
# Ejercicio 7: Torres de Hanoi
def torres_de_hanoi(n, origen, destino, auxiliar):
    if n == 1:
        print(f"Mover disco de {origen} a {destino}")
        return 1

    movimientos = torres_de_hanoi(n - 1, origen, auxiliar, destino)

    print(f"Mover disco de {origen} a {destino}")
    movimientos += 1

    movimientos += torres_de_hanoi(n - 1, auxiliar, destino, origen)

    return movimientos

# Pruebas

print("--- Prueba con 2 discos ---")
total_2 = torres_de_hanoi(2, 'A', 'C', 'B')
print(f"Total movimientos: {total_2}")

print("\n--- Prueba con 3 discos ---")
total_3 = torres_de_hanoi(3, 'A', 'C', 'B')
print(f"Total movimientos: {total_3}")
```

```python
# Ejercicio 8: MCD recursivo 
def mcd(a, b):
    if b == 0:
        return a
    else:
        return mcd(b, a % b)
    
# Pruebas
print(f"mcd(48, 18)  -> {mcd(48, 18)}")  # Debería dar 6
print(f"mcd(100, 50) -> {mcd(100, 50)}") # Debería dar 50
print(f"mcd(17, 19)  -> {mcd(17, 19)}")  # Debería dar 1
```

```python
# Ejercicio 9: Comparar eficiencia
import timeit

def fib_recursivo(n):
    if n <= 1:
        return n
    return fib_recursivo(n - 1) + fib_recursivo(n - 2)


def fib_iterativo(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

n_test = 30

tiempo_rec = timeit.timeit(lambda: fib_recursivo(n_test), number=5) / 5

tiempo_ite = timeit.timeit(lambda: fib_iterativo(n_test), number=5) / 5

# Medir tiempos y comparar
print(f"Resultados para Fibonacci({n_test}):")
print(f"Recursivo: {tiempo_rec:.6f} segundos")
print(f"Iterativo: {tiempo_ite:.6f} segundos")

diferencia = tiempo_rec / tiempo_ite
print(f"\nLa versión iterativa es aproximadamente {int(diferencia):,} veces más rápida.")
```
